```python
import pandas as pd

url = 'https://www.officialcharts.com/chart-news/the-best-selling-albums-of-all-time-on-the-official-uk-chart__15551/'
albums = pd.read_html(url, header = 0)[0]
albums
```

```python
albums = albums.rename(columns = {'TITLE':'TYTUŁ', 'ARTIST':'ARTYSTA', 'YEAR': 'ROK', 'HIGH POSN' : 'MAX POZ'})
albums
```

```python
unique_artists = albums['ARTYSTA'].nunique()
unique_artists
```

```python
albums['ARTYSTA'].value_counts()
```

```python
albums = albums.rename(str.title, axis = 1)
albums
```

```python
albums = albums.drop('Max Poz', axis = 1)
albums
```

```python
albums['Rok'].value_counts()
```

```python
chosen_years = albums.groupby((albums['Rok'] < 1990) & (albums['Rok'] > 1960)).count()
chosen_years.iloc[1][0]
```

```python
albums['Rok'].min()
```

```python
first_albums = albums.sort_values(by = 'Rok').drop_duplicates(subset='Artysta')
first_albums
```

```python
first_albums.to_csv('first_albums_csv', index=False)
```
