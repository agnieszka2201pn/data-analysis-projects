```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
movies = pd.read_csv('tmdb_movies.csv')
genres = pd.read_csv('tmdb_genres.csv')
```

```python
movies
```

```python
quantile3 = movies.quantile(q=0.75)
quantile3['vote_count']
```

```python
movies_quantile3 = movies[movies['vote_count'] > quantile3['vote_count']]
```

```python
movies_by_vote_average = movies_quantile3.sort_values('vote_average', ascending=False)
best10movies = movies_by_vote_average[:10]
best10movies
```

```python
movies['release_date'] = pd.to_datetime(movies['release_date'])
movies2010_2016 = movies[(movies['release_date'] >= '2010-01-01') & (movies['release_date'] <= '2016-12-31')]
revenue_budget_mean = movies2010_2016.groupby(pd.Grouper(key = 'release_date', freq = 'Y')).agg({'revenue' : 'mean', 'budget' : 'mean'})
revenue_budget_mean.reset_index(inplace = True)
revenue_budget_mean
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.set_title('Mean revenue and budget of movies produced in 2010 - 2016')
revenue_budget_mean['revenue'].plot(kind = 'bar', label = 'revenue')
revenue_budget_mean['budget'].plot(label = 'budget', color = 'red')
axes.legend(loc= (1.05, 0.9))
def format_y(n,pos):
    return f'{n/1000000}M'
formatter = plt.FuncFormatter(format_y)
axes.yaxis.set_major_formatter(formatter)
axes.set_xticklabels(x for x in range(2010, 2017))
```

Conclusions: It can be observed that average budget spent on movies production in years 2010-2013 was on a steady level. In 2014-2015 there was a certain decline in the spends, followed by sharp growth in 2016. At the same time revenues have been growing slowly but steadily, with one peak in 2012 and another one in 2016. It can be concluded that the increase of spends does not necessarily cause immediate increase of the revenues. Further reaserch would be needed to understand the correlation between movie budget and revenues and influence of different factors on both values.

```python
genres_dict = dict(zip(genres['Unnamed: 0'], genres['genres']))
movies['genre_name'] = movies['genre_id'].map(genres_dict)
movies
```

```python
movies_by_genre = movies['genre_name'].value_counts()
movies_by_genre
```

```python
runtime_mean = movies.groupby(pd.Grouper(key='genre_name')).agg({'runtime':'mean'})
runtime_mean.reset_index()
runtime_mean.sort_values(by = 'runtime', ascending = False).iloc[0]
```
