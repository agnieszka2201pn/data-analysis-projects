```python
from math import sqrt

file = 'Book1.csv'
with open(file) as file_object:
    lines = file_object.readlines()
    
fertility_rates = []
life_expectancy = []
    
for line in lines:
    fertility_rates.append(float(line.rstrip().split(',')[1]))
    life_expectancy.append(float(line.rstrip().split(',')[2]))
    
#print(life_expectancy, fertility_rates)
    
def average(input_list):
    average_from_the_sample = sum(input_list)/len(input_list)
    
    #for the whole population
    lower_limit = round(average_from_the_sample-1.96*(9.09/sqrt(len(input_list))),2)
    upper_limit = round(average_from_the_sample + 1.96*(9.09/sqrt(len(input_list))),2)
    result = (lower_limit + upper_limit)/2
        
    return f'Average from the sample = {average_from_the_sample}.\nLife expectancy for the whole population = {result}'

def median(input_list):
    input_list_sorted = sorted(input_list)
    if len(input_list) % 2 != 0:
        result = input_list_sorted[int(((len(input_list)-1))/2)]
    else:
        result = sum(input_list_sorted[int(len(input_list)/2-1): (int(len(input_list)/2)+1)])/2
        
    return f'Median from the sample = {result}'

def dominanta(input_list):
    dominanta = None
    for n in range(len(input_list)):
        if input_list.count(input_list[n]) > input_list.count(input_list[n-1]):
            dominanta = input_list[n]
    return f'Dominant in the sample = {dominanta}'


def variance_and_sd(input_list):
    average = sum(input_list)/len(input_list)
    sum_of_squares = 0
    for n in range(len(input_list)):
        sum_of_squares += (input_list[n]-average) ** 2
    result = round(sum_of_squares/(len(input_list)-1), 2)
    return f'Variance in the sample = {result}, standard deviation = {round(sqrt(result),2)}'


def correlation(input_list1, input_list2):
    average_input_list1 = sum(input_list1)/len(input_list1)
    average_input_list2 = sum(input_list2)/len(input_list2)
    
    #covariance
    component_sum = 0
    for n in range(len(input_list1)):
        component_sum += ((input_list1[n]-average_input_list1)*(input_list2[n]-average_input_list2))
    covariance = component_sum/(len(input_list1)-1)
    
    
    #stdev1
    sum_of_squares_list1 = 0
    for n in range(len(input_list1)):
        sum_of_squares_list1 += (input_list1[n]-average_input_list1) ** 2
    stdev1 = sqrt(sum_of_squares_list1/(len(input_list1)-1))
    
    #stdev2
    sum_of_squares_list2 = 0
    for n in range(len(input_list2)):
        sum_of_squares_list2 += (input_list2[n]-average_input_list2) ** 2
    stdev2 = sqrt(sum_of_squares_list2/(len(input_list2)-1))
    

    correlation = round(covariance/(stdev1*stdev2), 2)
    
    return f'Correlation coefficient = {correlation}'



print(average(life_expectancy))
print(median(life_expectancy))
print(dominanta(life_expectancy))
print(variance_and_sd(life_expectancy))
print(correlation(fertility_rates, life_expectancy))
                    
    
```
